Clustering

K-means clustering

```python
from sklearn.cluster import KMeans
import pandas as pd

# Load the student performance dataset
df = pd.read_csv("Student_Performance_dataset.csv")

# Drop rows with missing values
df.dropna(inplace=True)

# Define the features for clustering
features = ['Age', 'English_Score', 'Math_Score', 'Science_Score']

# Fit a KMeans model to the data
model = KMeans(n_clusters=3)  # You can change the number of clusters as needed
model.fit(df[features])

# Visualize the results
import matplotlib.pyplot as plt

# Predict cluster labels
y_kmeans = model.predict(df[features])

# Scatter plot the features
plt.scatter(df['Age'], df['English_Score'], c=y_kmeans, s=50, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('English Score')
plt.title('K-means Clustering of Student Performance')
plt.show()
```

Choosing number of clusters k

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import pandas as pd

# Load the student performance dataset
df = pd.read_csv("Student_Performance_dataset.csv")

# Drop rows with missing values
df.dropna(inplace=True)

# Define the features for clustering
features = ['Age', 'English_Score', 'Math_Score', 'Science_Score']

# Extract feature data
X = df[features]

# Trying different number of clusters
for k in range(2, 11):
    model = KMeans(n_clusters=k).fit(X)
    label = model.labels_
    sil_coeff = silhouette_score(X, label, metric='euclidean')
    print("For n_clusters={}, The Silhouette Coefficient is {:.3f}".format(k, sil_coeff))
```

Anomaly detection with k means

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd

# Load the student performance dataset
df = pd.read_csv("Student_Performance_dataset.csv")

# Drop rows with missing values
df.dropna(inplace=True)

# Define the features for clustering
features = ['Age', 'English_Score', 'Math_Score', 'Science_Score']

# Extract feature data
X = df[features]

# Fit K-means model with only one cluster
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)

# Visualize the results
import matplotlib.pyplot as plt

# Scatter plot the data points
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title('K-means clustering with only one cluster')
ax.scatter(X['Age'], X['English_Score'], c='blue', s=50, label='Points')

# Plot the centroid
centroid = kmeans.cluster_centers_[0]
ax.scatter(centroid[0], centroid[1], c='red', s=200, alpha=0.5, label='Centroid')

ax.legend()
plt.xlabel('Age')
plt.ylabel('English Score')
plt.show()
```

```python
# Calculate distances of each point to the centroid
distances = kmeans.transform(X)

# Sort the distances and get the indexes of the top five anomalous points
sorted_idx = np.argsort(distances.ravel())[::-1][:5]

# Visualize the data points along with the centroid and top five anomalous points
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title("Anomaly Detection")
ax.scatter(X['Age'], X['English_Score'], label='Points')

# Plot the centroid
centroid = kmeans.cluster_centers_[0]
ax.scatter(centroid[0], centroid[1], label='Centroid', color='r')

# Plot the top five anomalous points
ax.scatter(X.iloc[sorted_idx]['Age'], X.iloc[sorted_idx]['English_Score'],
           label='Anomalies', color='y', s=100)

ax.legend(loc='best')
plt.xlabel('Age')
plt.ylabel('English Score')
plt.show()
```

Removing anomaly points

```python
# Convert sorted_idx to a regular Python list
sorted_idx_list = sorted_idx.tolist()

# Remove the anomalous points from the original dataset
new_X = np.delete(X, sorted_idx_list, axis=0)

# Extract the anomalies
anomaly_X = X[sorted_idx_list]
print('Coordinates of the five anomalies:')
print(anomaly_X)


```
